```python
#Define numerical and categorical feature
numerical_features=[feature for feature in df.columns if df[feature].dtype!='O']
categorical_features=[feature for feature in df.columns if df[feature].dtype=='O']

# print columns
print("We have {} numerical Feature :{}".format(len(numerical_features),numerical_features))
print("We have {} categorical Feature :{}".format(len(categorical_features),categorical_features))
```

# Adding column for total score and average score

```python
df.head(2)
```

```python
df['Total_Score']=df['math_score']+df['reading_score']+df['writing_score']
df['Avg_score']=df['Total_Score']/3
df
```

```python
reading_full=df[df['reading_score']==100]['Avg_score'].count()
writing_full=df[df['writing_score']==100]['Avg_score'].count()
math_full=df[df['math_score']==100]['Avg_score'].count()
print(f"Number of Student with full marks in Math :{math_full}")
print(f"Number of Student with full marks in Writing :{writing_full}")
print(f"Number of Student with full marks in Reading :{reading_full}")
```

```python
reading_less20=df[df['reading_score']<=20]['Avg_score'].count()
writing_less20=df[df['writing_score']<=20]['Avg_score'].count()
math_less20=df[df['math_score']<=20]['Avg_score'].count()
print(f'Number of student less than 20 marks in  Reading :{reading_less20} ')
print(f'Number of student less than 20 marks in  Writing :{writing_less20} ')
print(f'Number of student less than 20 marks in  Math :{math_less20} ')
```

## Exploring Data {Visulisation}

```python
fig,axs=plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='Avg_score',bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='Avg_score',kde=True,hue='gender')
plt.show()
```

```python
fig,axs=plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='Total_Score',bins=30,kde=True,color='g')
plt.subplot(122)
sns.histplot(data=df,x='Total_Score',kde=True,hue='gender')
plt.show()
```

```python
fig,axs=plt.subplots(1,3,figsize=(25,6))
plt.subplot(141)
sns.histplot(data=df,x='Avg_score',bins=30,kde=True,hue='lunch')
plt.subplot(142)
sns.histplot(data=df[df.gender=='female'],x='Avg_score',kde=True,hue='lunch')
plt.subplot(143)
sns.histplot(data=df[df.gender=='male'],x='Avg_score',kde=True,hue='lunch')
plt.show()
```
